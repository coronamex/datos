```python
import pycurl
import json
from io import BytesIO
import pandas as pd
```

```python
url = "http://ncov.sinave.gob.mx/mapa.aspx/Grafica22"
```

```python
buffer = BytesIO()
c = pycurl.Curl()
c.setopt(c.URL, url)
c.setopt(c.WRITEDATA, buffer)
data = json.dumps({})
c.setopt(pycurl.POST, 1)
c.setopt(pycurl.POSTFIELDS, data)
c.setopt(pycurl.HTTPHEADER,[
#          'User-Agent: Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:74.0) Gecko/20100101 Firefox/74.0',
#          'Accept: */*',
#          'Accept-Language: es-ES,es;q=0.8,en-US;q=0.5,en;q=0.3',
         'Content-Type: application/json; charset=utf-8',
#          'X-Requested-With: XMLHttpRequest',
#          'Origin: http://ncov.sinave.gob.mx',
#          'Connection: keep-alive',
#          'Referer: http://ncov.sinave.gob.mx/mapa.aspx',
#          'Cookie: acceptcookiefreecounterstat=ok; counter=7c1b9c58284212977601e1056d7ecf31; counter_nv=7c1b9c58284212977601e1056d7ecf31',
#          'DNT: 1',
#          'Cache-Control: max-age=0'
])
c.perform()
c.close()

body = buffer.getvalue()
mapa_json = body.decode('UTF-8')
# print(body.decode('UTF-8'))
tab = json.loads(json.loads(mapa_json)['d'])
tab
```

```python
df = pd.DataFrame(tab,
                  columns=['id', 'estado', 'pob', 'num',
                               'casos_acumulados',
                                'negativos_acumulados',
                               'sospechosos', 'muertes_acumuladas'],
                 dtype='int')
# df.sort_values('casos_acumulados')
df = df.drop(columns = ['id','pob', 'num'])
# df.sort_values('casos_acumulados')
ii = ['casos_acumulados',
                                'negativos_acumulados',
                               'sospechosos', 'muertes_acumuladas']
df[ii] = df[ii].apply(pd.to_numeric)
df.sort_values('muertes_acumuladas')
# df
```
